```python
from dataclasses import dataclass
import ipywidgets as widgets
from ruamel.yaml import YAML

import functools
from IPython.core.display import HTML, SVG
display(HTML("<style>.container { width:90% !important; }</style>"))
import os
import sys
import glob
sys.path.append("..")

%load_ext autoreload
%autoreload 2

yaml = YAML()
```

```python
config_path = os.getenv("CONFIG_FILE_PATH", f"config/config.yaml")
project_path = os.getenv("PROJECT_PATH", f"/home/fxlyonnet/repos/shape-ce/examples/test")

with open(os.path.join(project_path, config_path), "r") as config_fd:
        config = yaml.load(config_fd)
```

```python
glob_react = glob.glob(os.path.join(project_path, config["results_dir"], "figures", config["folders"]["reactivity"], "*.svg"))
glob_normreact = glob.glob(os.path.join(project_path, config["results_dir"], "figures", config["folders"]["normreact"], "*.svg"))
glob_aggreact = glob.glob(os.path.join(project_path, config["results_dir"], "figures", config["folders"]["aggreact"], "*.svg"))
glob_varna = glob.glob(os.path.join(project_path, config["results_dir"], config["folders"]["varna"], "*.svg"))
glob_footprint = glob.glob(os.path.join(project_path, config["results_dir"], config["folders"]["footprint"], "*.svg"))
```

```python
display(HTML("<h3>Reactivity</h3>"))

for svg in glob_react[:1]:
    display(SVG(url="file://"+svg))
    
display(HTML("<h3>Normalized Reactivity</h3>"))

for svg in glob_normreact[:1]:
    display(SVG(url="file://"+svg))
    
display(HTML("<h3>Aggregated Reactivity</h3>"))

glob_aggreact.sort()
for svg in glob_aggreact[:2]:
    display(SVG(url="file://"+svg))

display(HTML("<h3>Structure</h3>"))
    
glob_varna.sort()
for svg in glob_varna[:1]:
    display(SVG(url="file://"+svg))


display(HTML("<h3>Footprint</h3>"))
    
glob_footprint.sort()
for svg in glob_footprint[:1]:
    display(SVG(url="file://"+svg))
    
```
